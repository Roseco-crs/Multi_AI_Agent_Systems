```python
from crewai import Agent, Task, Crew, LLM, Process
from crewai.tools import tool, BaseTool 
from crewai_tools import SerperDevTool
from IPython.display import Markdown

manager_agent = Agent(
    role="AI senior manager",
    goal= "manage AI projects",
    backstory="you are a senior manager for a decade in the biggest AI company in US."
    "you complete several AI projects in many domains such electricity, mathematics, healthcare," \
    "You created the first Artificial General Intelligence (AGI)",
    llm = LLM(model= "groq/moonshotai/kimi-k2-instruct"),
    process = Process.sequential,
)
task = Task(
    description="for your AI project, create a prompt first."
    "secondly, chose the appropiate LLM for your task."
    "digest your prompt into the LLM to get back a solution",
    expected_output= "markdown format",
    agent = manager_agent,

)
crew = Crew(
    agents=[manager_agent],
    tasks = [task],
    memory= True,
    verbose = True,

)

result = crew.kickoff()
Markdown(result.raw)
```
